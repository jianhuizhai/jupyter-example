Here I will try also equations, plots, maths and later also BInder integration.

- square area = $(2 r)^2$
- circle area = $\pi r^2$
- circle / square = $\pi r^2 / 4 r^2$ = $\pi / 4$
- $\pi$ = 4 * (circle/square)

![Darts](https://coderefinery.github.io/jupyter/img/darts.svg)

Here I import the "random" module to get random numbers

```python
import random
```

initialize variables

```python
N = 10000
points = []
```

Here we throw darts

```python
hits = 0
for i in range(N):
    x, y = random.random(), random.random()
    if x**2 + y**2 < 1.0:
        hits += 1
        points.append((x, y, True))
    else:
        points.append((x, y, False))

```

plot results

```python
%matplotlib inline
from matplotlib import pyplot
x, y, colors = zip(*points)
pyplot.scatter(x, y, c=colors)

```

Approximation for $\pi$

```python
fraction = hits / N
4 * fraction
```
